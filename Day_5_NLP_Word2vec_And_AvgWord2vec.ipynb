```python
# importing the Dataset

import pandas as pd

messages = pd.read_csv('/content/SMSSpamCollection.txt', sep='\t',
                           names=["label", "message"])
```

```python
messages
```

```python
messages['message'].iloc[451]
```

```python
#Data cleaning and preprocessing
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['message'][i])
    review = review.lower()
    review = review.split()

    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
# Creating the Bag of Words model
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500)
X = cv.fit_transform(corpus).toarray()
```

```python
print(X)
```

```python
y=pd.get_dummies(messages['label'])
y=y.iloc[:,1].values
```

```python
y
```

```python
# Train Test Split

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_detect_model = MultinomialNB().fit(X_train, y_train)
```

```python
#prediction
y_pred=spam_detect_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score=accuracy_score(y_test,y_pred)
print(score)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_pred,y_test))
```

```python
import numpy as np

def avg_word2vec(doc, model):
    # Filter words in the document to ensure they exist in the model
    valid_words = [word for word in doc if word in model.wv.index_to_key]

    if not valid_words:  # Check if there are valid words
        # If no valid words, return a zero vector or some default vector
        return np.zeros(model.vector_size)  # Return a 100-dimensional zero vector

    # Calculate the average vector for the valid words
    return np.mean([model.wv[word] for word in valid_words], axis=0)
```

```python
# Apply for the entire sentences
X = []
for i in tqdm(range(len(words))):
    print("Processing sentence index:", i)
    avg_vector = avg_word2vec(words[i], model)
    X.append(avg_vector)
```
